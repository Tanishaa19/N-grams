```python
import re
from collections import Counter
```

```python
with open("/content/newsdata.txt", "r", encoding='latin-1') as file:
    text = file.read()
```

```python
text = re.sub(r'[^\w\s]', '', text.lower())
tokens = text.split()
unigrams = [(tokens[i],) for i in range(len(tokens))]
bigrams = [(tokens[i], tokens[i+1]) for i in range(len(tokens)-1)]
trigrams = [(tokens[i], tokens[i+1], tokens[i+2]) for i in range(len(tokens)-2)]

print("\n Unigrams:\n ", unigrams)
print("\n Bigrams:\n", bigrams)
print("\n Trigrams:\n", trigrams)
```

```python
bigram_freq = Counter(bigrams)
print("\n Repeating Bigrams")
for gram, count in bigram_freq.items():
    if count > 1:
        print(gram, count)
```

```python
import matplotlib.pyplot as plt
top = bigram_freq.most_common(5)
x = [' '.join(i[0]) for i in top]
y = [i[1] for i in top]
plt.bar(x, y)
plt.show()
```
